```python
import pandas as pd
data= pd.read_csv('mammographic_masses.data.txt',na_values=["?"], names=['BI-RADS', 'age', 'shape', 'margin', 'density', 'severity'])
data.head()
```

```python
data.loc[(data['age'].isnull()) | (data['shape'].isnull()) | (data['margin'].isnull()) | (data['density'].isnull())]
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
data.head()
```

```python
features= data[["age", 'shape', 'margin', 'density']]. values
feature_names=['age', 'shape', 'margin', 'density']
```

```python
labels= data[['severity']].values
```

```python
print(labels.shape)
print(features.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(features)
features= scaler.transform(features)
features
```

```python
import numpy as np
np.random.seed(1234)
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(features, labels, test_size= 0.25, random_state=1)
```

```python
from sklearn import tree
clf= tree.DecisionTreeClassifier(random_state=1)
clf.fit(xtrain, ytrain)
```

```python
clf.score(xtest,ytest
         )
```

```python
from sklearn.model_selection import cross_val_score
model= tree.DecisionTreeClassifier()
model.fit(features, labels)
```

```python
cv_scores= cross_val_score(model, features, labels, cv=10)
print(cv_scores)
print(cv_scores.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=10, random_state=1)
scores= cross_val_score(clf, features, labels, cv=10)
scores.mean()
```

```python
from sklearn import svm

C = 1.0
svc = svm.SVC(kernel='linear', C=C)
cv_scores = cross_val_score(svc, features, labels, cv=10)

cv_scores.mean()
```

```python
from sklearn import neighbors
m=0
n=0
for i in range(1, 50):
    clf = neighbors.KNeighborsClassifier(n_neighbors=i)
    score= cross_val_score(clf, features, labels, cv=10)
    mean= scores.mean()
    if mean>m:
        m=mean
        n=i
    else:
        pass
print(n, m)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
transformed_features= scaler.fit_transform(features)
clf= MultinomialNB()
scores= cross_val_score(clf, transformed_features, labels, cv=10)
print(scores.mean())
```

```python
from sklearn.linear_model import LogisticRegression
clf= LogisticRegression()
scores= cross_val_score(clf, features, labels, cv=10)
print(scores.mean())
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

def create_model():
    model = Sequential()
    model.add(Dense(6, input_dim=4, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier


estimator = KerasClassifier(build_fn=create_model, epochs=10, verbose=2)

cv_scores = cross_val_score(estimator, features, labels, cv=10)
cv_scores.mean()
```
